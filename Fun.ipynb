```python
def uniques(x):
    list1=[]
    for i in x:
        list1+=i 
    list2=list1.copy()    
    for item in list2:
        if list2.count(item) >1:
            list2.remove(item)
    c=[]        
    for u in list2:
        c.append(list1.count(u))
    return c
    
```

```python
def roman_numerals():
    rom_num = input('enter the roman numeral: '.upper()).lower()
    if 4 in uniques(rom_num):
        print('invalid roman numeral!!'.upper())
    else:    
        tot = roman_num_dict.get(rom_num[0])
        if len(rom_num)>1:
            for z in rom_num[1:]:
                t=roman_num_dict.get(z)
                if tot > t or tot==t:
                    tot=tot+t
                elif tot < t:
                    tot=t-tot
        else:
            tot=roman_num_dict.get(rom_num)
        return tot       
```

```python
def Sequences_series(sequence,arithmetic=False,geometric=False):
    if arithmetic==True:
        a=sequence[0]
        d= sequence[1]-a
        print(f'a = {a}')
        print(f'd = {d}')
        choice=input('enter operation: u for nth term or s for sum: '.upper())
        if choice=='u':
            n=int(input('enter the nth number: '.upper()))
            print(a+(n-1)*d)
        else:
            n=int(input('enter number of first terms to be added: '.upper()))
            print(n/2*(2*a+(n-1)*d))
    if geometric==True:
        a=sequence[0]
        r=sequence[1]/sequence[0]
        print(f'a = {a}')
        print(f'r = {r}')
        
        choice=input('enter operation: u for nth term or s for sum: '.upper())
        if choice=='u':
            n=int(input('enter the nth number: '.upper()))
            print(a*r**(n-1))
        else:
            n=int(input('enter number of first terms to be added: '.upper()))
            if r<1:
                print(a*((1-(r**n)/(1-r))))
            else:
                print(a*(((r**n)-1)/(r-1)))
        
        
```

```python
def unique(s):
    for i in s:
        if s.count(i)>1:
            s.remove(i)
    return s        
```

```python
def root_finder(a=0,b=0,c=0,d=0,e=0,f=0,p=0,q=0):
    listr=[]
    for i in range(-p,p+1):
        if i!=0:
            for j in range(-q,q+1):
                if j!=0:
                    listr.append(i/j)
    listr=unique(listr)
    for x in listr:
        if (a*x**5 + b*x**4 + c*x**3 + d*x**2 + e*x + f)==0 :
            print(x)
```
